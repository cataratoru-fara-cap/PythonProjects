```python
def create_df(path: str, nrows: int = None, skiprows: int = None) -> pd.DataFrame: 
    df = pd.read_csv(filepath_or_buffer=path, nrows=nrows, skiprows=skiprows).transpose()

    # reasign labels
    new_labels = df.iloc[0]
    df = df.drop(df.index[0]) 
    df.columns = new_labels 

    #cleans Unamend: 7
    df = df[:-1] 
    df.columns.name = None
    
    return df
```

```python
b1 = create_df("b).csv", nrows=6, skiprows=None)
b2 = create_df("b).csv", nrows=6, skiprows=8)
b3 = create_df("b).csv", nrows=6, skiprows=15)

b3
```

Désigné, sur le même graphique, les caractéristiques de sortir 

```python
plt.title('les caractéristiques de sortir Ici = f(VCEi) i= 1,2,3')

plt.plot(b1["I_C1(mA)"], b1["V_CE (V)"], marker="x")
plt.plot(b2["I_C2(mA)"], b2["V_CE (V)"], marker="x")
plt.plot(b3["I_C3(mA)"], b3["V_CE (V)"], marker="x")

plt.gca().set_ylabel('V', rotation=0)
plt.xlabel('mA')
plt.grid(True)
```
