```python
import numpy as np
import matplotlib.pyplot as plt

mu = 350 #mean
sigma = 16.667 #standard deviation
zaruri = np.random.normal(mu, sigma, 10000000)
```

```python
plt.hist(zaruri, linewidth=2,
       label='normal', bins = 1000)
plt.xlabel('Total')
plt.ylabel('Procentaj')
plt.title('Probabilitatea Zarurilor')
plt.show()
```

```python
zaruri = np.random.binomial(100 ,1/6 ,size = 10000)
```

```python
plt.hist(zaruri,linewidth=2,
       label='normal', bins=50)
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.show()
```
